```python
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd
import matplotlib.dates as mdates


# Plotting with adjusted closing prices
plt.figure(figsize=(20, 10))
plt.plot(tesla['Adj Close'])  # Use 'Adj Close' for adjusted prices

# Set x-axis limits
plt.xlim(pd.Timestamp('2010-06-29'), pd.Timestamp('2024-09-20'))


# Format the x-axis to show year-month-day
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator(1))  # Set tick interval to 1 year



# Add labels and title
plt.title('Tesla Adjusted Close Prices (Log Scale)')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price (Log Scale)')
plt.grid(True)

# Show the plot
plt.show()
```

**Time Resample**

```python
tesla.resample(rule='A').min()
```

```python
#year and frequency
tesla.resample(rule='A').min().plot()
```

```python
tesla.resample(rule='A').max()
```

```python
tesla.resample(rule='A').mean()
```

```python
#quarterly start frequency
#year and frequency
tesla.resample(rule='Q').min()['Open'].plot()
```

```python
#business
tesla.resample(rule='BM').max()['Open'].plot()
```

```python
tesla.resample(rule='A').max()['Open'].plot(kind='bar')
```
